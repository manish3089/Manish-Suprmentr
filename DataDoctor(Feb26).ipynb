```python
import pandas as pd
data = pd.read_csv("messy_data.csv")
df = pd.DataFrame(data)
```

```python
df.sample(5)
```

```python
# Step 1 : Understand the data
df.info()
```

```python
df.shape
```

```python
df.columns
```

```python
# Step 2: Handle duplicates
df.duplicated(subset=["Name","Email","City","Order_Amount","Order_Date"])
```

```python
df.drop_duplicates(subset=["Name","Email","City","Order_Amount","Order_Date"])
```

```python
# Step 3: Standardizing text data

df['City']=df['City'].str.upper()
```

```python
df
```

```python
# Step 4: Handle Missing Values
''' NaN - Not a number
None
empty cells'''
df.isnull()
```

```python
df.isnull().sum()
```

```python
# remove missing values
df.dropna()
```

```python
df.dropna(axis=0)
# axis = 0 -> rows
# axis = 1 -> columns
```

```python
df['Order_Amount'] = pd.to_numeric(df['Order_Amount'],errors='coerce')
df['Order_Date'] = pd.to_datetime(df['Order_Date'],errors='coerce')
```

```python
df.info()
```

```python
df
```

```python
# fill missing values
# for numerical columns ->mean/median
df['Order_Amount'].fillna(df['Order_Amount'].mean(),inplace=True)
```

```python
# filling missing values in categorical columns

df['Email'].fillna(df['Name'].str.split().str[0].str.lower() + '@gmail.com',inplace = True)
```

```python
df.dropna(inplace = True)
```
