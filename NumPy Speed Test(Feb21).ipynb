```python
import random
import numpy as np
my_list = [random.randint(1,1000000) for _ in range (1000000)]
```

```python
my_array = np.array(my_list)
my_array
```

```python
import time

start = time.time()
element_wise_list = [x*2 for x in my_list]
end = time.time()
print("List time:", end - start)

start = time.time()
element_wise_array = my_array *2
end = time.time()
print("Numpy time:", end - start)
```

```python
start = time.time()
addition_list = sum(my_list)
end = time.time()
print("List time:", end - start)

start = time.time()
addition_numpy = np.sum(my_array)
end = time.time()
print("Numpy time:", end - start)
```

```python
import math

start = time.time()
sqr_list = [math.sqrt(x) for x in my_list]
end = time.time()
print("List time:", end - start)

start = time.time()
sqr_numpy = np.sqrt(my_array)
end = time.time()
print("Numpy time:", end - start)
```

🔹 Which operation gave the biggest speedup?
- Numpy



🔹 Did NumPy’s advantage grow with more complex math (like sqrt) compared to simple multiplication?
- Yes.

Multiplication is already fast in both lists and NumPy, but NumPy still wins because it avoids Python’s per-element loop overhead.

Square root/exponential operations show an even bigger gap because NumPy calls optimized C/Fortran routines under the hood, while lists rely on Python’s math.sqrt for each element, which is slower due to repeated function calls.
- Observation: The more complex the math, the more NumPy’s vectorization advantage compounds.

🔹 In what scenarios would you still use Python lists?
Heterogeneous data → lists can store mixed types (strings, numbers, objects).

Dynamic structures → lists allow easy insertion, deletion, resizing.

Small-scale tasks → for a few hundred elements, the difference in speed is negligible, and lists are simpler.
- Observation: Lists are flexible and general-purpose, while NumPy arrays are specialized for numerical, large-scale computations.
