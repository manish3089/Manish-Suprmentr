```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

iris = load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['species']= pd.Series(iris.target).map(dict(enumerate(iris.target_names)))
df.sample(5)
```

```python
for col in numerical_cols:
    sns.histplot(df[col],bins=50,kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
df['species'].value_counts()
```

```python
def feature_stats(df):
    for col in numerical_cols:
        feature_data = df[col]
        feature_mean = feature_data.mean()
        feature_median = feature_data.median()
        feature_std = feature_data.std()
        feature_count = feature_data.count()
    
        print(f"{col}")
        print(f"Mean:{feature_mean:.2f}")
        print(f"Median:{feature_median:.2f}")
        print(f"Std:{feature_std:.2f}")
        print(f"Count:{feature_count}")
        print()
feature_stats(df)
```
