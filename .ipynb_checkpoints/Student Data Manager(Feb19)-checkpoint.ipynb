```python
import numpy as np
Students = {"Manish":50,
"Asif ": 49,
"Hasan" : 48,
"Naman" : 40,
"Theja": 38
}
topper = max(Students, key = Students.get)
average = np.mean(list(Students.values()))
```

```python
def assign_grade(score):
    if score >= 90:
        return "A+"
    elif score >= 75:
        return "A"
    elif score >= 60:
        return "B"
    elif score >=45:
        return "C"
    elif score >=35:
        return "D"
    else:
        return "F"

n = int(input("Enter the number of students: "))

for i in range(n):
    name = input(f"Enter the name of student: ")
    score = int(input(f"Enter the score of {name}: "))
    Students[name] = score

for name,score in Students.items():
    grade = assign_grade(score)
    print(f"{name} scored {score} -> Grade: {grade}")
```
